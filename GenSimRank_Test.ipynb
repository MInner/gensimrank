```python
import numpy as np
from scipy.sparse import csc_matrix
```

```python
class GenType:    
    def __init__(self, name):
        self.name = name
        self.__elements = []
        
    def getid(self, element):
        try:
            return self.__elements.index(element)
        except:
            self.__elements.append(element)
            return len(self.__elements) - 1
        
    def objbyid(self, id):
        return self.__elements[id]
        
    def size(self):
        return len(self.__elements)
    
    def elements(self):
        return self.__elements
    
    def __repr__(self):
        return "<GenType %s has %s>" % (self.name, self.__elements)
        
class GenRelation:
    
    def __init__(self, name, types, symmetric = True):
        if not (isinstance(types[0], GenType) and isinstance(types[1], GenType)):
            raise ValueError, "excpect types to be (GenType, GenType)"
        self.name = name
        self.typeFrom = types[0]
        self.typeTo = types[1]
        self.rowones = []
        self.colones = []
        self.symmetric = symmetric
    
    def add(self, a_name, b_name):
        self.rowones.append(self.typeFrom.getid(a_name))
        self.colones.append(self.typeTo.getid(b_name))            
                            
    def __repr__(self):
        return "<GenRelation '%s' between '%s' and '%s': %s>" % (
                self.name, 
                self.typeFrom.name, 
                self.typeTo.name, 
                [(self.typeFrom.objbyid(id_from), self.typeTo.objbyid(id_to)) for id_from, id_to in zip(self.rowones, self.colones)]
            )
    
    def tomatrix(self):
        if self.typeTo == self.typeFrom and self.symmetric:
            N, M = self.typeFrom.size(), self.typeTo.size()
            return csc_matrix((np.ones(2*len(self.rowones)), (self.rowones + self.colones, self.colones + self.rowones)), shape = (N, M))

        N, M = self.typeFrom.size(), self.typeTo.size()
        return csc_matrix((np.ones(len(self.rowones)), (self.rowones, self.colones)), shape = (N, M))

def unzero(x):
    r = x[:]
    r[x==0] = 1
    return r   
    
def row_normalized(A):
    return (A.T.astype(float) / unzero(A.T.sum(axis=0))).T

class GenSimRankModel:
    
    def __init__(self):
        self.types = {}
        self.relations = {}
        
    def getTypeByName(self, name):
        try:
            return self.types[name]
        except:
            raise ValueError, "No such type %s"%name
    
    def getRelByName(self, name):
        try:
            return self.relations[name]
        except:
            raise ValueError, "No such relation %s"%name
    
    def addType(self, name):
        if not name in self.types.keys():
            t = GenType(name)
            self.types[name] = t
            return t
        else:
            raise ValueError, "Type %s already exists in this model" % name
            
    def batchAddTypes(self, typenameslist):
        for tn in typenameslist:
            self.addType(tn)
            
    def addRelation(self, name, fr_name, to_name, symmetric = True):            
        if not name in self.relations.keys():
            fr_t = self.getTypeByName(fr_name)
            to_t = self.getTypeByName(to_name)            
            r = GenRelation(name, [fr_t, to_t], symmetric)
            self.relations[name] = r
            return r
        else:
            raise ValueError, "Relation %s already exists in this model" % name
        
    def buildOrderRelation(self, gentype_name):
        gentype = self.getTypeByName(gentype_name)
        sortedelements = sorted(gentype.elements())
        followedBy = self.addRelation('followedBy_' + gentype_name, gentype_name, gentype_name, symmetric = False)
        for fr, to in zip(sortedelements, sortedelements[1:]):
            followedBy.add(fr, to)
        
        return followedBy
    
    def add(self, relation_name, fr, to):
        r = self.getRelByName(relation_name)
        r.add(fr, to)
        
    def buildRelationBlocks(self):
        types = self.types.values()
        blocks = []
        w = np.zeros((len(types), len(types)))
        for r in self.relations.values():
            t1_id = types.index(r.typeFrom)
            t2_id = types.index(r.typeTo)
            W = r.tomatrix()
            blocks.append((t1_id, t2_id, row_normalized(W) ))
            w[t1_id, t2_id] = 1
            if t1_id != t2_id:
                blocks.append((t2_id, t1_id, row_normalized(W.T)))
                w[t2_id, t1_id] = 1
                
        if np.any(w.T.sum(axis=0) < 0.5):
            raise ValueError, "Graph is not connected! Wont converge. See types: %s \n %s" % ([t.name for i, t in enumerate(types) if i in np.where(w.T.sum(axis=0) < 0.5)], w) 
                
        return blocks, types, row_normalized(w)
    
```

__Conclusion:__

Iterative process
$$ S_i^{(k+1)} = \sum_j w_{ij}  W^{(i,j)} S_j^{(k)} W^{(j,i)} - diag(\sum_j w_{ij} W^{(i,j)} S_j^{(k)} W^{(j,i)}) + I $$
converges if:

- $M^{(j,i)} = (M^{(i,j)})^T$
- $M^{(i,i)} = (M^{(i,i)})^T$
- $\forall i \sum_j w_{ij} = 1$
- $W^{(i,j)}_{lm} = \frac {M^{(i,j)}_{lm}} {\sum_p M^{(i,j)}_{lp}} \iff \forall (i, j, l) \sum_m W^{(i,j)}_{lm} = 1$

Note: if some row is empty (no connection) - doesn't converge.
